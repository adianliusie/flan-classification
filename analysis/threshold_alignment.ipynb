```python
import matplotlib.pyplot as plt

dataset = 'rt'
path = f"/home/al826/rds/hpc-work/2023/prompts/calib/search/submit/outputs/flan-t5-large/{dataset}/prompt_0"

fig, ax = plt.subplots(figsize=(6,5.5))

logits_reader = LogitsReader(path=path, dataset=dataset)
logits_reader.threshold_plot(norm='log-norm', ax=ax, color='blue')
logits_reader.threshold_plot(norm='null-norm', ax=ax, color='red')

plt.xlabel(r'$\frac{P_{\theta}(w_1|p)}{P_{\theta}(w_2|p)}$', size=18)
plt.ylabel(r'$\tau_{optimal}$', size=18)

plt.legend(['marginalisation', 'null'])
```
